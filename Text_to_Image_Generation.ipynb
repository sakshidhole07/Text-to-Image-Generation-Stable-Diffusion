## Step 2: Import Required Libraries

Here we import libraries for:
- Using GPU
- Loading Stable Diffusion model
- Handling generated images


```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
```

## Step 3: Load Pre-trained Stable Diffusion Model

This step loads the already trained Stable Diffusion model
and moves it to GPU for faster image generation.


```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

## Step 4: Provide Text Prompt

A text prompt describes what kind of image
the AI model should generate.


```python
prompt = "A futuristic smart city at night with glowing skyscrapers, flying cars, neon lights, artificial intelligence controlling traffic, ultra-realistic, cinematic lighting, 4K resolution"
```

## Step 5: Generate Image from Text Prompt

In this step, the AI model converts the given
text prompt into an image.


```python
image = pipe(prompt).images[0]
```

## Step 6: Display and Save Generated Image

The generated image is displayed and saved
for final submission.


```python
from IPython.display import display

display(image)
image.save("generated_image_2.png")
```
